```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
movies = movies.merge(credits,on='title')
movies.head()
```

```python
movies.info()
```

```python
movies = movies.drop(["homepage"],axis=1)
```

```python
movies.info()
```

```python
movies = movies.drop(["tagline"],axis=1)
```

```python
movies.info()
```

```python
movies = movies.dropna()
movies.info()
```

```python
movies.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
movies = movies[["title","genres","popularity","keywords","revenue","vote_count","vote_average","cast","crew","overview"]]
```

```python
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
import ast
def stos(stro):
    s = []
    for i in ast.literal_eval(stro):
        s.append(i['name'])
    return s
```

```python
movies.head()
```

```python
movies.iloc[0].genres
```

```python

movies['genres'] = movies['genres'].apply(stos)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(stos)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(stos)
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
movies['crew'] = movies['crew'].apply(fetch_director)

movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
import nltk
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
new_df.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
vector = cv.fit_transform(new_df['tags']).toarray()
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
        return " ".join(y)
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python


from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
sns.heatmap(similarity)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)),reverse = True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('Batman Begins')
```
